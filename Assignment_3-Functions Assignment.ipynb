```python
# 3. What are the different ways to define and call a function in Python?
#Ans: In Python, functions can be defined and called in several ways:
# Normal function (without arguments)
def greet():
    print("Hello")
greet()   # calling function

# Function with arguments
def add(a, b):
    return a + b
print(add(5, 7))

# Function with default arguments

def greet(name="D.J"):
    print("Hello", name)
greet()         # uses default
greet("Deepesh")  # overrides default


# Function with variable-length arguments
# *args → handles multiple positional arguments
# **kwargs → handles multiple keyword arguments

def show(*args, **kwargs):
    print(args)
    print(kwargs)

show(1, 2, 3, name="Deepesh", age=22)

# Lambda function (anonymous function)

square = lambda x: x * x
print(square(5))
```

```python
# 6. Explain the concept of generators in Python and how they are defined.
# Ans:Generators are special type of function which uses yeild instead of return.
# It provides value one at a time.Generators don't store values instead  generates value at a fly and iti is memory efficient
# example:
def count_to(n):
    i = 1
    while i <= n:
        yield i   # yields one value at a time
        i += 1

gen=count_to(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
# Ans: Map():
# The map() function executes a specified function for each item in an iterable. The item is sent to the 
# function as a parameter
# Reduce ():
# Reduce function implements a mathematical technique known as folding or reduction
# It is a process where you reduce a list of items to a single value.
# Filter():
# Filter is a built-in function used to filter elements from an iterable (like list, tuple) based on a condition.
# Example:
# Map():
num=[2,3,4,5]
add=list(map(lambda x:x+x, num))
print(add)
# Reduce
from functools import reduce
num=[1,2,3,4]
mul=reduce(lambda x,y:x*y, num)
print(mul)
# Filter()
num=[2,3,4,5,6,7,8,9,10]
fil=list(filter(lambda x:x%2==0,num))
print(fil)
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.
# Ans:
def  num(arg):
    nums=0
    for i in arg:
        if i%2==0:
            nums+=i
    print(nums)

num([2,3,4,5,6])        
        
```

```python
# 4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:  
        return False
    for i in range(2, int(n**0.5) + 1):  
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num)

```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
# Ans:
def fib(n):
    a=0
    b=1
    for i in range (n):
        yield a
        a,b=b,a+b
        
list(fib(8))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_two(n):
    for i in range(n+1):
        yield 2**i
for i in power_two(5):
    print(i)
        
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
# Ans:In this question I took help with chat-gpt.
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
for line in read_lines("read.txt"):
    print(line)
```
